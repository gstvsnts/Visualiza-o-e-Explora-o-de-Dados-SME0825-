```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style = 'darkgrid')
```

```python
data = {
    'Tipo de Eletrodoméstico' : ['Condicionadores de ar',
                                 'Secadores de roupas',
                                 'Máquinas de lavar',
                                 'Computadores',
                                 'Forno elétrico',
                                 'Lava-Louças',
                                 'Freezers',
                                 'Iluminação',
                                 'Refrigeradores',
                                 'Aquecedores de ambiente',
                                 'Aquecedores de água',
                                 'Televisores e decodificadores'],
    'Porcentagem %': [18,5,24,1,2,2,2,16,9,7,8,6]
    }
```

```python
df = pd.DataFrame(data)
df
```

```python
fig, ax = plt.subplots(figsize=(17,9))
ax = sns.barplot(df, y = 'Porcentagem %', x = 'Tipo de Eletrodoméstico')
ax.set_title('Consumo de Energia Elétrica residencial nos Estados Unidos', fontsize = 24, pad = 15)
plt.xticks(rotation = 70)
plt.yticks([0,5,10,15,20])
ax.set_ylabel('%', rotation = 180, fontsize = 16, loc = 'top')
ax.set_xlabel('')
```

```python
df2 = df.sort_values(by = 'Porcentagem %', ascending = False)
df2 = df2.reset_index()
df2 = df2.drop('index', axis = 1)
```

```python
df2['Acumulado'] = df2['Porcentagem %'].cumsum()
```

```python
df2
```

```python
fig, ax = plt.subplots(figsize=(17,9))

ax.bar(x = df2['Tipo de Eletrodoméstico'], height = df2['Porcentagem %'])
ax.set_title('Gráfico de Pareto (%)', fontsize = 24, pad = 15)
plt.xticks(rotation = 70)

ax2 = ax.twinx()
ax2.plot(df2['Tipo de Eletrodoméstico'], df2['Acumulado'], color = 'red', marker = '+')
plt.grid(visible = False)
ax2.set_ylabel('Acumulado', loc = 'top')
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.pie(x = df2['Porcentagem %'], labels = df2['Tipo de Eletrodoméstico'])
ax.set_title('Gráfico de Pizza', fontsize = 24)
plt.xticks(rotation = 60)
```
